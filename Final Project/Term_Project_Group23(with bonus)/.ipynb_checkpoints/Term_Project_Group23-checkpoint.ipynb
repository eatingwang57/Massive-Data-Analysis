```python
from pyspark import SparkContext, SparkConf, SparkFiles
sc.stop()
conf = SparkConf().set("spark.default.parallelism", 4).set('spark.driver.maxResultSize', '10G')
sc =  SparkContext(conf=conf)
```

```python
# PART1
# mapper1
def readFile(line):
    result = line.split(',')
    if (not result[0].isdigit()):
        return
    else:
        return int(result[1]), [[int(result[0]), result[2]]]
        # movieID, [userID, rating]

def reducer1(a, b):
    return a+b

def Calculate(line):
    # line = (movieID, [userID, rating])
    total = 0
    subMean = []
    subMeanSqr = 0
    merge = []
    for person in line[1]:
        total += float(person[1])
    mean = total / len(line[1])
    
    for person in line[1]:
        subMean.append(float(person[1]) - mean)
        subMeanSqr += pow(float(person[1]) - mean, 2)
    subMeanSqr = pow(subMeanSqr, 0.5)
    
    for i in range(len(line[1])):
        merge.append((line[1][i][0], [[line[0], subMean[i], subMeanSqr]]))
        # userID, [movieID, rating to this movie - mean, meanSquare]
    return merge



def Similarity(line):
    # line = userID, [ [movieID, rating-mean, meanSquare], [movieID, rating-mean, meanSquare], ... ]
    tmp = []
    for i in range(len(line[1])):
        for j in range(i + 1, len(line[1])):
            length = line[1][i][2] * line[1][j][2]
            if length == 0:
                tmp.append(((line[1][i][0], line[1][j][0]),[0]))
            else:
                tmp.append(((line[1][i][0], line[1][j][0]), [line[1][i][1] * line[1][j][1] / length]))
            # (moveID1, moveID2), [sim])
    return tmp


def Add(line):
    tmp = 0
    for sim in line[1]:
        tmp += sim
    return line[0], tmp


# PART2
movie = 0

def readAgain(line):
    result = line.split(',')
    if (not result[0].isdigit()):
        return
    else:
        return int(result[0]), [[int(result[1]), result[2]]]
        # userID, [movieID, rating]
        
def findPredict(line):
    predict = []
    haveRating = []
    haveRating_id = []
    flag = False
    for item in line[1]:
        haveRating_id.append(item[0])
        haveRating.append(item)
    for i in range(1, movie+1): # movie = number of movie
        if i not in haveRating_id:
            predict.append(i)
     
    return line[0], [predict, haveRating_id, haveRating]
    # userID, [[movies that need to be predicted], [predicted]]
    

def Predict(line):
    # line = (userID, [(to be predicted, [(rating, sim), (rating, sim), ...]), (...)])
    tmp = []
    n = 0
    d = 0
    for predict in line[1]:
        movie = predict[0]
        for rating in predict[1]:
            if rating[1] > 0:
                n += rating[0] * rating[1]
                d += rating[1]
        if d > 0:
            tmp.append(((line[0], movie), n / d))
    return tmp
```

```python
path = "smallData.csv"
file = sc.textFile(path)

m1 = file.map(readFile).filter(lambda x: x != None).reduceByKey(reducer1).sortByKey(True)

m2 = m1.flatMap(Calculate).reduceByKey(reducer1)
m3 = m2.flatMap(Similarity).reduceByKey(reducer1)
```

```python
m3 = m3.map(Add).sortByKey(True)
# ((movieID1, movieID2), sim)
result_sim = m3.collect()
outputfile = open("OutputFile_small.txt", "w")
for i in range(len(result_sim)):
    outputfile.write("(" + str(result_sim[i][0][0]) + ', ' + str(result_sim[i][0][1]) + ') : ' + str(result_sim[i][1]))
    outputfile.write('\n')
outputfile.close()


#end of basicï¼ˆPART 1)
```

```python
m4 = file.map(readAgain).filter(lambda x: x != None).reduceByKey(reducer1).sortByKey(True)
movie = m1.count()

m5 = m4.map(findPredict)
m5 = m5.collect()
# userID, [predict, haveRating_id, haveRating]

array = []
for line in m5:
    data = []
    
    for movie in line[1][0]:
        # m3 = ((movieID1, movieID2), sim)
        m6 = m3.filter(lambda x: (x[0][0] == movie and x[0][1] in line[1][1]) or \
                       (x[0][1] == movie and x[0][0] in line[1][1])).sortBy(lambda x: -x[1])
        m6 = m6.take(10)

        tmp = []
        for pair in m6:
            for r in line[1][2]:
                rating = []
                if r[0] == pair[0][0] or r[0] == pair[0][1]:
                    tmp.append((float(r[1]), pair[1]))
                    break
        data.append((movie, tmp))
    array.append((line[0], data))

m7 = sc.parallelize(array)
# (userID, [(to be predicted, [(rating, sim), (rating, sim), ...]), (...)])
m8 = m7.flatMap(Predict)

result = m8.collect()
outputfile = open("OutputFile.txt", "w")
for i in range(len(result)):
    outputfile.write("(" + str(result[i][0][0]) + ', ' + str(result[i][0][1]) + ') : ' + str(result[i][1]))
    outputfile.write('\n')
outputfile.close()
```
